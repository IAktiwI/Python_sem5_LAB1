```python
import numpy as np
import csv

def get_column_data(filename, column_name):
    with open(filename, 'r') as file:
        reader = csv.DictReader(file)
        return np.array([int(row[column_name]) for row in reader if row[column_name].isdigit()])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Додаємо імпорт matplotlib для побудови графіків

# Завантажуємо дані
data = pd.read_csv("C:\\Users\\Systema\\OneDrive\\Рабочий стол\\Завдання(5)\\Обробка даних Python\\Лаб №1\\russia_losses_equipment.csv")

# Видаляємо зайві пробіли в назвах стовпців
data.columns = data.columns.str.strip()

# Перетворюємо стовпець Date в тип datetime
data['date'] = pd.to_datetime(data['date'])

# Перевіряємо, чи відсортовані дані за датою
data = data.sort_values(by='date')  # Сортуємо дані за датою

# Перетворюємо дані для APC (можливо, що потрібно створити масив добових втрат)
apc_data = data['APC'].to_numpy()  # Зберігаємо значення APC у масив
daily_losses_apc = np.diff(apc_data)  # Обчислюємо добові втрати

# Додаємо добові втрати до DataFrame (зміщуємо їх на 1 рядок вниз)
data['Daily_losses_APC'] = np.insert(daily_losses_apc, 0, 0)  # Вставляємо 0 на перше місце

# Вибираємо останні 200 днів
last_200_days = data.tail(200)

# Створюємо графік
plt.figure(figsize=(16, 8))
plt.plot(last_200_days['date'], last_200_days['Daily_losses_APC'], linestyle='--', color='blue', label="Втрати APC")
plt.title("Втрати APC за останні 200 днів")
plt.xlabel("Дата")
plt.ylabel("Кількість втрат")
plt.legend()
plt.grid(color='gray', linestyle=':', linewidth=0.5)
plt.xticks(rotation=45)
plt.tight_layout()

# Зберігаємо графік у файл PNG
plt.savefig("apc_losses_last_200_days.png", dpi=100)

# Виводимо графік
plt.show()
```
